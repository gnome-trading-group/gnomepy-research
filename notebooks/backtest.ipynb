# Backtest

End-to-end example of running a `gnomepy` backtest against a strategy. PnL is marked-to-market on every market tick, so the curve moves with price (not just on fills).

```python
%load_ext autoreload
%autoreload 2

import os

os.environ.setdefault("GNOME_ROOT", "~/src/gnome")

from gnomepy.java._jvm import ensure_jvm_started
ensure_jvm_started()
```

```python
from datetime import datetime

from gnomepy import Backtest, ExchangeConfig, SchemaType
from gnomepy_research.strategies.momentum import MomentumTaker
from gnomepy_research.reporting import BacktestReport
```

```python
from gnomepy_research.presets import load_preset

backtest = load_preset(
    "mm_btc_30m",
    start_date=datetime(2026, 1, 23, 10, 30),
    end_date=datetime(2026, 1, 23, 13, 0),
)
```

```python
results = backtest.run()
```

```python
report = BacktestReport(results, backtest=backtest)
report.summary_df()
```

```python
report.plot_pnl().show()
```

```python
report.plot_position().show()
```

```python
from gnomepy_research.reporting import ReportSection, market_making_section, adverse_selection_section

report.save_html("report.html",
  extra_sections=[
      ReportSection("adverse", "Adverse Selection", adverse_selection_section),
      ReportSection("mm", "Market Making", market_making_section),
  ],
)
```
